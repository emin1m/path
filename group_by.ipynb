#  we are grouping the data by "preferred_foot" and "overall_rating" and then calculating the mean of the "crossing" and "potential" columns for each group.

# Normalize the "overall_rating" within each preferred_foot group using z-score


```python
def normalize_zscore(x):
    return (x - x.mean()) / x.std()

data['overall_rating_normalized'] = data.groupby('preferred_foot')['overall_rating'].transform(normalize_zscore)

# Print the result
print(data[['preferred_foot', 'overall_rating', 'overall_rating_normalized']])
```

```python
filtered_data = data[data['overall_rating_normalized'] > 1]
filtered_data
```

# Calculate the difference between the max and min overall_rating for each preferred_foot group

```python
def max_min_diff(x):
    return x.max() - x.min()
result = data.groupby('preferred_foot')['overall_rating'].apply(max_min_diff)

print(result)
```

```python
def cal_average(x):
    crossing_avg = x['crossing'].mean()
    poteantial_avg= x['potential'].mean()
    return pd.Series({'crossing_avg':crossing_avg, 'poteantial_avg': poteantial_avg})

result = data.groupby('preferred_foot').apply(cal_average)

result
```
