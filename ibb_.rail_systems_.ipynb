```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df=pd.read_excel('rayl-sistemler-gunluk-aylkyllk-hat-bazl-sefer-saylar.xlsx')
df
```

```python
df.dtypes
```

### First 17 rows are important for us

```python
df = df.iloc[:18].copy()
df
```

```python
yearly_totals = df.groupby('Yıl')[['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M9', 'T1', 'T4', 'T5', 'F1', 'T3', 'TF1', 'TF2']].sum()
yearly_totals.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Total Passengers by Year')
plt.xlabel('Year')
plt.ylabel('Total Passengers')
plt.show()

```

```python
monthly_averages = df.groupby('Periyot')[['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M9', 'T1', 'T4', 'T5', 'F1', 'T3', 'TF1', 'TF2']].mean()
monthly_averages.T.plot(kind='bar', figsize=(12, 6))
plt.title('Monthly Averages Over the Years')
plt.xlabel('Month')
plt.ylabel('Average Passengers')
plt.legend(title='Periyot', bbox_to_anchor=(1, 1))
plt.show()
```

```python
total_passengers_per_line = df[['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M9']].sum()
total_passengers_per_line.plot(kind='bar', figsize=(8, 6))
plt.title('Total Passengers for Each Line')
plt.xlabel('Line')
plt.ylabel('Total Passengers')
plt.show()
```

```python
df_M1= df[['Yıl', 'Periyot', 'M1']].copy()
df_M1
```

```python
df_M1_gunluk = df_M1[df['Periyot'] == 'Günlük']
df_M1_gunluk
```

```python
df_M1_gunluk.plot(kind='scatter',
                 x='Yıl',
                 y='M1',
                 title= 'Average Daily Passenger on M1')
plt.show()
```

```python
ax = sns.scatterplot(x='Yıl',
                y='M1',
                hue='Periyot',
                data=df)
ax.set_title('Passenger Details on M1')
plt.show()
```

```python
m1_m2_data = df[['M1', 'M2']]
correlation_coefficient = m1_m2_data['M1'].corr(m1_m2_data['M2'])
plt.figure(figsize=(8, 6))
sns.scatterplot(data=m1_m2_data, x='M1', y='M2', hue=m1_m2_data.index, palette='coolwarm', legend='full')
plt.title('Correlation between M1 and M2')
plt.xlabel('M1')
plt.ylabel('M2')
plt.grid(True)
plt.text(400, 700, f'Correlation Coefficient: {correlation_coefficient:.2f}', fontsize=12, bbox=dict(facecolor='white', edgecolor='black'))
plt.show()
```

```python
m1_data = df[df['Periyot'] == 'Günlük'][['M1']]

# Create a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=m1_data, y='M1')
plt.title('Box Plot for Line M1 (Daily Passenger Count)')
plt.ylabel('Passengers')
plt.grid(True)
plt.show()
```

```python
m_columns = df.filter(like='M', axis=1)

# Calculate the correlation matrix
correlation_matrix = m_columns.corr()

# Create the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for M Columns')
plt.show()
```

### Analyz last 5 year where there is no NaN in dataset

```python
df_new = df.iloc[12:18].copy()
df_new
```

```python
m_columns = df_new.filter(like='M', axis=1)

# Calculate the correlation matrix
correlation_matrix = m_columns.corr()

# Create the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for M Columns')
plt.show()
```

```python
correlation_m1_t1 = df_new['M1'].corr(df_new['T1'])

colors = ['blue', 'green']

sns.scatterplot(data=df_new, x=df_new.index, y='M1', color=colors[0], marker='o', alpha=0.7, label='M1')
sns.scatterplot(data=df_new, x=df_new.index, y='T1', color=colors[1], marker='o', alpha=0.7, label='T1')

plt.title(f"Correlation between M1 and T1: {correlation_m1_t1:.2f}")
plt.xlabel('Index')
plt.ylabel('Passengers')
plt.legend()

plt.grid(True)
plt.show()
```
