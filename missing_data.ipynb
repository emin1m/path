```python
space_ship.dropna() # Will drop entire row or coloumn if there is NaN in a row or in a coloumn
```

```python
space_ship.dropna(subset= ['RoomService']) # only drop the NaN values in the RoomService coloumns
```

```python
space_ship.dropna(how="all") #will drop the row if all  record or values are NaN in the  that row
```

```python
space_ship.dropna(how="any") #will drop row if any  record or values are NaN in the row == same as .dropna()
```

```python
space_ship.dropna(thresh=1) # there should be atleast 1 value in that row if true then row will not drop
```

```python
space_ship.dropna(thresh=13) #there should be atleast 13 value in that row if true then row will not drop
```

```python
space_ship.dropna(thresh=14) #there should be atleast 14 value in that row if true then row will not drop == same as .dropna()
```

### fillna method, instead of dropping down we will fill with a different values or text

```python
 space_ship.fillna("Value is missing") #not very ideal cuz we deal with numbers to analyze the data 
```

```python
space_ship.fillna(method = "ffill") # Forward-fill missing values with the last valid value.
```

```python
space_ship.fillna(method = "bfill") # Backward-fill missing values with the next valid value.
```
